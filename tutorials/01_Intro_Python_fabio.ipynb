### Creating GUIs in a few seconds
One of the great features of the notebook is to create super easily small interactive GUIs that are super helpful to test/debug a function.

````python
from ipywidgets import interact
def plotx(a,alpha,b,title='',ylog=False):
    x=np.linspace(0,100,num=1e3)
    y=a*x**alpha+b
    plt.plot(x,y)
    if ylog: plt.yscale('log')
    plt.title(title)
    plt.show()

interact(plotx,a=(1,10),alpha=(0.5,2,0.25),b=(0,5))    
````
